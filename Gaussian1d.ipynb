Here is a random draw from such a survey with reasonable parameters:

```python
np.random.seed(0)

muX = 0.0
sigmaX = 1.0
xth = .5


sigma_obs = 0.5

N = 100
Ntot = 0
Nobs = 0 
#np.random.poisson(100)

xs_all = []

xs_obs_all =[]



xs_obs = []


while(Nobs<N):



    x = np.exp(muX + sigmaX*np.random.randn())
    xo = np.exp(np.log(x) + sigma_obs*np.random.randn())

    
    if (xo > xth):
        
        xs_obs.append(xo)

    
        
        Nobs+=1
        
    xs_obs_all.append(xo)

  
    
    xs_all.append(x)

```

And a comparison between the distribution of true $x$s (green) and the observed $x$s (blue).

```python
fig, ax = plt.subplots()

sns.distplot(np.log(xs_obs))
sns.distplot(np.log(xs_obs_all))
```

```python
model = pystan.StanModel(file='gaussian.stan')
```

```python
fit = model.sampling(
    data={'Nobs': len(xs_obs),
          'xobs': xs_obs,
          'sigma_obs': sigma_obs, 
          'xth': xth, 
          
          'NNobs_max': 100},
    iter=32000/2,
    thin=16/2,
    seed=123

)
```

```python
chain = fit.extract(permuted=True)
```

```python
fit.plot(['mu_x', 'sigma_x', 'Lambda']);
plt.tight_layout()
```
